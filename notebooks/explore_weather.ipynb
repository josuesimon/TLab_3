```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# We will load the data from our CSV file br_final.csv and display the first few rows of the data
data = pd.read_csv('../weather_api/data/csv/br_final.csv')
print("Data Overview:")
print(data.head())

# 1. The first graph produced will show the harvest outcomes in Minas Gerais over the year
minas_data = data[data['subdivision'] == 'Minas Gerais']
plt.figure(figsize=(12, 6))
sns.lineplot(data=minas_data, x='year', y='million_60kgs_bag', marker='o', color='orange')
plt.title('Harvest Production in Minas Gerais Over the Years')
plt.xlabel('Year')
plt.ylabel('Coffee Production (Million 60kg Bags)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 2. The second graph will show the harvest outcomes by region in Brazil
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='year', y='million_60kgs_bag', hue='subdivision', marker='o')
plt.title('Harvest Production by Region in Brazil Over the Years')
plt.xlabel('Year')
plt.ylabel('Coffee Production (Million 60kg Bags)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 3. The third graph will show the total coffee production per each region in Brazil
top_producers = data.groupby('subdivision')['million_60kgs_bag'].sum().reset_index()
top_producers = top_producers.sort_values(by='million_60kgs_bag', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_producers, x='subdivision', y='million_60kgs_bag', color='orange')
plt.title('Total Coffee Production by Region in Brazil')
plt.xlabel('Region')
plt.ylabel('Total Coffee Production (Million 60kg Bags)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# 4. The fourth graph will show the average unemployment percentage per each region in Brazil
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='year', y='avg_unemp_perc', hue='subdivision', marker='o')
plt.title('Average Unemployment Percentage by Region in Brazil Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Unemployment Percentage')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Summary of Findings
print("""
1. While coffee production in Minas Gerais has been inconsistent from 2002-2023, growth has still been occuring despite the short falls that is faced in some years.
2. Minas Gerais is a strong competitor among other regions in Brazil, constantly being at the top of the production line.
3. Again, this graph shows how Minas Gerais is a top producer of coffee beans among other regions in Brazil.
4. Minas Gerais stands second in place in the unemploymant chart, showing a correlation between its strong production and its connection to offering employment to many of its people. Therefore helping keep the unemployment percentage down to compared to other regions which are not as competitive in production.
""")
```
